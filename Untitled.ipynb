```python
import pandas as pd
import numpy as np

# 1. Create a list of 5 timestamps (just strings for now)
timestamps = ['10:00', '10:01', '10:02', '10:03', '10:04']

# 2. Create a list of 5 log levels (Mix of 'INFO' and 'ERROR')
# TODO: Edit this list to have mostly 'INFO' and one 'ERROR'
log_levels = ['INFO', 'INFO', 'ERROR', 'INFO', 'INFO'] 

# 3. Create a list of 5 response times (integers)
# TODO: Edit this list. Make 4 normal (e.g., 50-100) and 1 high (e.g., 2000)
response_times = [50, 60, 70, 8000, 90] 

# 4. Combine them into a Dictionary
data = {
    'timestamp': timestamps,
    'log_level': log_levels,
    'response_time_ms': response_times
}

# 5. Create the DataFrame
df = pd.DataFrame(data)

# 6. Display the table
df
```

```python
# Filter 1: Find all ERROR logs
# TODO: Complete the condition inside the brackets
error_logs = df[df['log_level'] == 'ERROR']

# Filter 2: Find all high response time logs (> 1000ms)
# TODO: Complete the condition
high_latency = df[df['response_time_ms'] > 1000]

# Display results
print("Error Logs:")
print(error_logs)
print("\nHigh Latency Logs:")
print(high_latency)
```

```python
import numpy as np
import pandas as pd

# 1. Generate 100 normal response times (e.g., mean=100, std=20)
# TODO: Adjust loc and scale to match your definition of 'normal'
normal_times = np.random.normal(loc=100, scale=20, size=100)

# 2. Generate 5 anomalous response times (e.g., mean=2000, std=50)
# TODO: Adjust loc to be significantly higher than normal
anomaly_times = np.random.normal(loc=2000, scale=50, size=5)

# 3. Combine them
all_times = np.concatenate([normal_times, anomaly_times])

# 4. Create Labels (0 = Normal, 1 = Anomaly) for verification later
labels = [0]*100 + [1]*5

# 5. Create DataFrame
df_large = pd.DataFrame({
    'response_time_ms': all_times,
    'is_anomaly': labels
})

# 6. Inspect
print(f"Total Rows: {len(df_large)}")
print(df_large.head())
# TODO: Run this to count the labels
# Replace 'column_name' with the actual column name if different
print(df_large['is_anomaly'].value_counts())
```

```python
from sklearn.ensemble import IsolationForest

# 1. Prepare the data (model expects 2D array)
# TODO: Reshape your response_time_ms column to 2D
X = df_large[['response_time_ms']]  # Double brackets make it 2D

# 2. Initialize the model
# TODO: Set contamination to match your known anomaly rate (5/105 ≈ 0.05)
model = IsolationForest(contamination=0.05, random_state=42)

# 3. Train the model
# TODO: Fit the model on your data
model.fit(X)

# 4. Make predictions
# TODO: Predict on the same data
predictions = model.predict(X)

# 5. Add predictions to DataFrame
df_large['prediction'] = predictions

# 6. Inspect results
print(df_large.head(10))
# TODO: Filter to show only predicted anomalies
predicted_anomalies = df_large[df_large['prediction'] == -1]
print(predicted_anomalies)
```
